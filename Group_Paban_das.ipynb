```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the loan dataset
loan = pd.read_csv("loan.csv")
```

```python
#Reading top few lines on loan dataset to validate the dataset loaded successfully
loan.head()
```

```python
#DATA QUALITY CHECK
```

```python
#Checking the loan shape
loan.shape
```

```python
#Check the datatypes of all the columns of the dataframe
loan.info()
```

```python
#checking columns with null values
loan.isnull().sum()
```

```python
print(loan.isnull())
```

```python
#DATA CLEANING
```

```python
#Dropping rows where all values are missing
loan.dropna(how='all')
```

```python
loan.shape
```

```python
#Dropping columns with all or at least 1 null value
loan = loan.dropna(axis=1)
```

```python
#checking loan shape after dropping rows and columns with missing values
loan.shape
```

```python
#making sure no null values present
loan.isnull().sum()
```

```python
#Filtering for Defaulters from loan dataset
```

```python
#checking all loan status values
loan['loan_status'].unique
```

```python
#Filtering for loan status with value "Charged Off" to get only loan defaulters dataset
loan_default = loan[loan['loan_status'].str.contains('Charged Off')]
```

```python
loan_default.info
```

```python
loan_default.shape
```

```python
loan_default.columns
```

```python
#BI VARIATE ANALYSIS
```

```python
#comparing Loan amount with Term
sns.jointplot(x="term", y="loan_amnt", data=loan_default)
plt.show()
```

```python
#comparing Loan amount with Home Ownership
sns.violinplot(x="home_ownership", y="loan_amnt", data=loan_default)
plt.show()
```

```python
#comparing Loan amount with Int_rate
sns.jointplot(x="int_rate", y="loan_amnt", data=loan_default)
plt.show()
```

```python
#comparing Loan amount with Purpose
ax = sns.barplot(x="purpose", y="loan_amnt", data=loan_default)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
#comparing home ownership with int_rate
ay = sns.stripplot(x="home_ownership", y="int_rate", data=loan_default)
ay.set_yticklabels(ax.get_yticklabels(), rotation=40, ha="right")
plt.show
```

```python
#comparing home ownership with total_rec_late_fee
sns.barplot(x="home_ownership", y="total_rec_late_fee", data=loan_default)
plt.show()
```

```python
#comparing purpose with total_rec_late_fee
x = sns.barplot(x="purpose", y="total_rec_late_fee", data=loan_default)
x.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```
